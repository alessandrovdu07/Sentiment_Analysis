# *venezuela-complete-2019*

## Clean file

```python
from typing import ClassVar
import pandas as pd

#Retrieve csv file data

from google.colab import files
uploaded = files.upload()

import io
csv = pd.read_csv(io.BytesIO(uploaded['venezuela-complete-2019.csv']))
# Dataset is now stored in a Pandas Dataframe
```

```python
print(csv.isnull().sum())
```

```python
df = csv.fillna(0)
df=df.drop(columns="Location")
df=df.rename(columns={'Location.1':'Location','Verfied':'Verified'})
```

## Begin EDA

```python
df.head()
```

```python
df.dtypes
```

What are the most famous tweets?

```python
df.sort_values('Favorites',ascending = False).head(5)
```

```python
numbers = df.select_dtypes(include=['number'])
import seaborn as sns
import matplotlib.pyplot as plt
g = sns.PairGrid(numbers)
g.map(plt.scatter);
```

Which were the most retweeted?

```python
df.sort_values('Retweets',ascending = False).head(5)
```

People with the most tweets

```python
df['Screen Name'].value_counts().nlargest(5)
```

# CambioClimatico-complete-2022


## Clean file

```python
from typing import ClassVar
import pandas as pd

#Retrieve csv file data

from google.colab import files
uploaded = files.upload()

import io
excel = pd.read_excel(io.BytesIO(uploaded['CambioClimatico-complete-2022.xlsx']))
# Dataset is now stored in a Pandas Dataframe
```

```python
print(excel.isnull().sum())
df = excel.fillna(0)
df=df.drop(columns=["Unnamed: 16","Unnamed: 17","Unnamed: 18"])
```

```python
df.dtypes
```

## Begin EDA

What are the most famous tweets?

```python
df.sort_values('Favorites',ascending = False).head(5)
```

What is the language share among the tweets?

```python
df['Language'].value_counts().nlargest(5)
```

Which are the most retweeted tweets?

```python
df.sort_values('Retweets',ascending = False).head(5)
```
